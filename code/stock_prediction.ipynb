```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout
import os
import warnings
import matplotlib.dates as mdates
from datetime import datetime, timedelta
import nbformat

# Setting style for visualizations
plt.style.use('fivethirtyeight')
sns.set_style('whitegrid')
warnings.filterwarnings('ignore')

# Function to display basic information about the dataset
def explore_dataset_structure():
    # Check directory structure
    print("Exploring dataset structure...")
    
    # Check the symbols metadata file
    symbols_meta = pd.read_csv('..\data\symbols_valid_meta.csv')
    print(f"Total symbols in metadata: {len(symbols_meta)}")
    print("\nSample of symbols metadata:")
    print(symbols_meta.head())
    
    # Count stock and ETF files
    stock_files = os.listdir('..\data\stocks')
    etf_files = os.listdir('..\data\etfs')
    print(f"\nNumber of stock files: {len(stock_files)}")
    print(f"Number of ETF files: {len(etf_files)}")
    
    # Display some example stock files
    print("\nSample stock files:")
    for file in stock_files[:5]:
        print(f"- {file}")
    
    return symbols_meta

# Function to load and preprocess a stock
def load_and_preprocess_stock(ticker, folder='..\data\stocks'):
    file_path = f"{folder}/{ticker}.csv"
    df = pd.read_csv(file_path)
    
    # Convert Date to datetime
    df['Date'] = pd.to_datetime(df['Date'])
    
    # Sort by date
    df = df.sort_values('Date')
    
    # Create additional features for analysis
    df['Year'] = df['Date'].dt.year
    df['Month'] = df['Date'].dt.month
    df['Day'] = df['Date'].dt.day
    df['DayOfWeek'] = df['Date'].dt.dayofweek
    
    # Calculate daily returns
    df['Return'] = df['Adj Close'].pct_change()
    
    # Calculate moving averages
    df['MA7'] = df['Adj Close'].rolling(window=7).mean()
    df['MA20'] = df['Adj Close'].rolling(window=20).mean()
    df['MA50'] = df['Adj Close'].rolling(window=50).mean()
    df['MA200'] = df['Adj Close'].rolling(window=200).mean()
    
    # Calculate volatility (standard deviation of returns)
    df['Volatility'] = df['Return'].rolling(window=20).std()
    
    return df

# Function to perform exploratory data analysis
def perform_eda(df, ticker):
    print(f"\nPerforming EDA for {ticker}...")
    print(f"Date range: {df['Date'].min()} to {df['Date'].max()}")
    print(f"Number of trading days: {len(df)}")
    
    # Basic statistics
    print("\nBasic statistics for Adjusted Close price:")
    print(df['Adj Close'].describe())
    
    # Visualize the stock price history
    plt.figure(figsize=(14, 7))
    plt.plot(df['Date'], df['Adj Close'], color='blue', label=f'{ticker} Adj Close')
    plt.title(f'{ticker} Stock Price History', fontsize=16)
    plt.xlabel('Date', fontsize=12)
    plt.ylabel('Price ($)', fontsize=12)
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
    
    # Visualize moving averages
    plt.figure(figsize=(14, 7))
    plt.plot(df['Date'], df['Adj Close'], color='blue', alpha=0.6, label=f'{ticker} Adj Close')
    plt.plot(df['Date'], df['MA7'], color='red', label='7-Day MA')
    plt.plot(df['Date'], df['MA20'], color='green', label='20-Day MA')
    plt.plot(df['Date'], df['MA50'], color='orange', label='50-Day MA')
    plt.plot(df['Date'], df['MA200'], color='purple', label='200-Day MA')
    plt.title(f'{ticker} Stock Price with Moving Averages', fontsize=16)
    plt.xlabel('Date', fontsize=12)
    plt.ylabel('Price ($)', fontsize=12)
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
    
    # Visualize daily returns
    plt.figure(figsize=(14, 7))
    plt.plot(df['Date'], df['Return'], color='green', alpha=0.6)
    plt.axhline(y=0, color='red', linestyle='-', alpha=0.3)
    plt.title(f'{ticker} Daily Returns', fontsize=16)
    plt.xlabel('Date', fontsize=12)
    plt.ylabel('Return (%)', fontsize=12)
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
    
    # Monthly average price
    monthly_avg = df.groupby(['Year', 'Month'])['Adj Close'].mean().reset_index()
    monthly_avg['YearMonth'] = pd.to_datetime(monthly_avg['Year'].astype(str) + '-' + monthly_avg['Month'].astype(str))
    
    plt.figure(figsize=(14, 7))
    plt.plot(monthly_avg['YearMonth'], monthly_avg['Adj Close'], marker='o', linestyle='-')
    plt.title(f'{ticker} Monthly Average Price', fontsize=16)
    plt.xlabel('Date', fontsize=12)
    plt.ylabel('Average Price ($)', fontsize=12)
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
    
    return

# Function to prepare data for LSTM model
def prepare_lstm_data(df, feature='Adj Close', look_back=60, test_size=0.2):
    # Select the feature to predict
    data = df[feature].values.reshape(-1, 1)
    
    # Scale the data
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(data)
    
    # Split into train and test sets
    train_size = int(len(scaled_data) * (1 - test_size))
    train_data = scaled_data[:train_size]
    test_data = scaled_data[train_size - look_back:]
    
    # Create sequences for training
    X_train, y_train = [], []
    for i in range(look_back, len(train_data)):
        X_train.append(train_data[i - look_back:i, 0])
        y_train.append(train_data[i, 0])
    
    X_train, y_train = np.array(X_train), np.array(y_train)
    X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
    
    # Create sequences for testing
    X_test, y_test = [], []
    for i in range(look_back, len(test_data)):
        X_test.append(test_data[i - look_back:i, 0])
        y_test.append(test_data[i, 0])
    
    X_test, y_test = np.array(X_test), np.array(y_test)
    X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
    
    test_dates = df['Date'].values[train_size:train_size + len(y_test)]
    
    return X_train, y_train, X_test, y_test, scaler, test_dates

# Function to build and train LSTM model
def build_lstm_model(X_train, y_train, X_test, y_test, scaler, epochs=10, batch_size=32):
    # Build the LSTM model
    model = Sequential()
    model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], 1)))
    model.add(Dropout(0.2))
    model.add(LSTM(units=50, return_sequences=False))
    model.add(Dropout(0.2))
    model.add(Dense(units=25))
    model.add(Dense(units=1))
    
    # Compile the model
    model.compile(optimizer='adam', loss='mean_squared_error')
    
    # Train the model
    history = model.fit(
        X_train, y_train,
        epochs=epochs,
        batch_size=batch_size,
        validation_data=(X_test, y_test),
        verbose=1
    )
    
    # Make predictions
    train_predictions = model.predict(X_train)
    test_predictions = model.predict(X_test)
    
    # Inverse transform predictions
    train_predictions = scaler.inverse_transform(train_predictions)
    test_predictions = scaler.inverse_transform(test_predictions)
    actual_y_train = scaler.inverse_transform(y_train.reshape(-1, 1))
    actual_y_test = scaler.inverse_transform(y_test.reshape(-1, 1))
    
    # Calculate metrics
    train_rmse = np.sqrt(mean_squared_error(actual_y_train, train_predictions))
    test_rmse = np.sqrt(mean_squared_error(actual_y_test, test_predictions))
    test_mae = mean_absolute_error(actual_y_test, test_predictions)
    test_r2 = r2_score(actual_y_test, test_predictions)
    
    print(f"\nModel Evaluation:")
    print(f"Train RMSE: {train_rmse:.4f}")
    print(f"Test RMSE: {test_rmse:.4f}")
    print(f"Test MAE: {test_mae:.4f}")
    print(f"Test R: {test_r2:.4f}")
    
    return model, history, train_predictions, test_predictions, actual_y_train, actual_y_test, test_rmse

# Function to visualize model predictions
def visualize_predictions(df, test_dates, actual_y_test, test_predictions, ticker, test_rmse, model):
    # Create a DataFrame for visualization
    prediction_df = pd.DataFrame({
        'Date': test_dates,
        'Actual': actual_y_test.flatten(),
        'Predicted': test_predictions.flatten()
    })
    
    # Plot the results
    plt.figure(figsize=(16, 8))
    
    # Set background color
    ax = plt.gca()
    ax.set_facecolor('#f5f5f5')
    
    # Plot actual vs predicted
    plt.plot(prediction_df['Date'], prediction_df['Actual'], 'b-', linewidth=2, label='Actual Price')
    plt.plot(prediction_df['Date'], prediction_df['Predicted'], 'r--', linewidth=2, label='Predicted Price')
    
    # Add shaded area between actual and predicted
    plt.fill_between(prediction_df['Date'], 
                     prediction_df['Actual'], 
                     prediction_df['Predicted'], 
                     alpha=0.3, 
                     color='green' if test_rmse < 10 else 'orange')
    
    # Format x-axis
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
    plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=2))
    plt.gcf().autofmt_xdate()
    
    # Add grid and styling
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.title(f'{ticker} Stock Price Prediction (RMSE: ${test_rmse:.2f})', fontsize=20)
    plt.xlabel('Date', fontsize=14)
    plt.ylabel('Stock Price ($)', fontsize=14)
    plt.legend(loc='best', fontsize=12)
    
    # Add annotation for model accuracy
    accuracy_text = f"RMSE: ${test_rmse:.2f}"
    plt.annotate(accuracy_text, xy=(0.02, 0.95), xycoords='axes fraction', 
                 fontsize=12, bbox=dict(boxstyle="round,pad=0.3", fc='white', alpha=0.7))
    
    plt.tight_layout()
    plt.show()
    
    # Create scatter plot of actual vs predicted
    plt.figure(figsize=(10, 8))
    
    # Calculate correlation
    correlation = np.corrcoef(prediction_df['Actual'], prediction_df['Predicted'])[0, 1]
    
    # Create scatter plot
    plt.scatter(prediction_df['Actual'], prediction_df['Predicted'], alpha=0.5, color='blue')
    
    # Add diagonal line (perfect prediction)
    min_val = min(prediction_df['Actual'].min(), prediction_df['Predicted'].min())
    max_val = max(prediction_df['Actual'].max(), prediction_df['Predicted'].max())
    plt.plot([min_val, max_val], [min_val, max_val], 'r--', linewidth=2)
    
    plt.title(f'{ticker} Actual vs Predicted Stock Prices (Correlation: {correlation:.4f})', fontsize=16)
    plt.xlabel('Actual Price ($)', fontsize=12)
    plt.ylabel('Predicted Price ($)', fontsize=12)
    plt.grid(True, alpha=0.3)
    plt.annotate(f'Correlation: {correlation:.4f}', xy=(0.02, 0.95), xycoords='axes fraction', 
                 fontsize=12, bbox=dict(boxstyle="round,pad=0.3", fc='white', alpha=0.7))
    plt.tight_layout()
    plt.show()
    
    # Plot the prediction error over time
    prediction_df['Error'] = prediction_df['Actual'] - prediction_df['Predicted']
    prediction_df['AbsError'] = abs(prediction_df['Error'])
    
    plt.figure(figsize=(16, 8))
    plt.plot(prediction_df['Date'], prediction_df['Error'], color='purple', linewidth=1.5)
    plt.axhline(y=0, color='red', linestyle='-', alpha=0.3)
    plt.fill_between(prediction_df['Date'], prediction_df['Error'], 0, 
                    prediction_df['Error'] > 0, color='green', alpha=0.3)
    plt.fill_between(prediction_df['Date'], prediction_df['Error'], 0, 
                    prediction_df['Error'] < 0, color='red', alpha=0.3)
    plt.title(f'{ticker} Prediction Error Over Time', fontsize=16)
    plt.xlabel('Date', fontsize=12)
    plt.ylabel('Error ($)', fontsize=12)
    plt.grid(True, alpha=0.3)
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
    plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=2))
    plt.gcf().autofmt_xdate()
    plt.tight_layout()
    plt.show()
    
    # Create a forecasting visualization (forecasting next 30 days)
    last_sequence = df['Adj Close'].values[-60:].reshape(-1, 1)
    scaler = MinMaxScaler(feature_range=(0, 1))
    last_sequence_scaled = scaler.fit_transform(last_sequence)
    
    X_forecast = last_sequence_scaled.reshape(1, 60, 1)
    
    # Create dates for forecast
    last_date = df['Date'].iloc[-1]
    forecast_dates = [last_date + timedelta(days=i) for i in range(1, 31)]
    forecast_dates = [date for date in forecast_dates if date.weekday() < 5]  # Only include weekdays
    
    # Create forecast
    forecast_values = []
    current_sequence = last_sequence_scaled.reshape(1, 60, 1)
    
    for _ in range(len(forecast_dates)):
        # Get the prediction (next day)
        next_pred = model.predict(current_sequence)
        # Add the prediction to our results
        forecast_values.append(next_pred[0, 0])
        # Update the sequence with the predicted value
        next_pred_reshaped = next_pred.reshape(1, 1, 1)
        current_sequence = np.append(current_sequence[:, 1:, :], next_pred_reshaped, axis=1)
    
    # Inverse transform
    forecast_values = scaler.inverse_transform(np.array(forecast_values).reshape(-1, 1))
    
    # Create forecast DataFrame
    forecast_df = pd.DataFrame({
        'Date': forecast_dates,
        'Forecasted': forecast_values.flatten()
    })
    
    # Plot the forecast
    plt.figure(figsize=(16, 8))
    
    # Plot historical data (last 90 days)
    historical_days = 90
    historical_df = df.iloc[-historical_days:]
    
    plt.plot(historical_df['Date'], historical_df['Adj Close'], 'b-', linewidth=2, label='Historical Price')
    plt.plot(forecast_df['Date'], forecast_df['Forecasted'], 'r--', linewidth=2, label='Forecasted Price')
    
    # Add shaded area for forecast
    plt.axvspan(last_date, forecast_df['Date'].iloc[-1], alpha=0.2, color='gray')
    
    # Format x-axis
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
    plt.gca().xaxis.set_major_locator(mdates.WeekdayLocator(interval=2))
    plt.gcf().autofmt_xdate()
    
    # Add grid and styling
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.title(f'{ticker} 30-Day Stock Price Forecast', fontsize=20)
    plt.xlabel('Date', fontsize=14)
    plt.ylabel('Stock Price ($)', fontsize=14)
    plt.legend(loc='best', fontsize=12)
    
    # Add annotation for last known price
    last_price = df['Adj Close'].iloc[-1]
    last_date_str = last_date.strftime('%Y-%m-%d')
    plt.annotate(f'Last Known Price (${last_price:.2f} on {last_date_str})', 
                 xy=(last_date, last_price),
                 xytext=(last_date - timedelta(days=5), last_price * 1.05),
                 arrowprops=dict(facecolor='black', shrink=0.05, width=1.5, headwidth=8),
                 fontsize=12, bbox=dict(boxstyle="round,pad=0.3", fc='white', alpha=0.7))
    
    plt.tight_layout()
    plt.show()
    
    return prediction_df, forecast_df

# Main function
def main():
    # Explore dataset structure
    symbols_meta = explore_dataset_structure()
    
    # Choose a stock to predict (AAPL as example)
    ticker = 'AAPL'
    
    # Get stock info from metadata
    stock_info = symbols_meta[symbols_meta['Symbol'] == ticker]
    if not stock_info.empty:
        print(f"\nSelected stock: {ticker} - {stock_info['Security Name'].values[0]}")
    
    # Load and preprocess stock data
    df = load_and_preprocess_stock(ticker)
    
    # Display the data
    print("\nStock data sample:")
    print(df.head())
    
    # Perform exploratory data analysis
    perform_eda(df, ticker)
    
    # Prepare data for LSTM model
    look_back = 60  # 60 days of history to predict the next day
    X_train, y_train, X_test, y_test, scaler, test_dates = prepare_lstm_data(df, look_back=look_back)
    
    print(f"\nPrepared data shapes:")
    print(f"X_train shape: {X_train.shape}")
    print(f"y_train shape: {y_train.shape}")
    print(f"X_test shape: {X_test.shape}")
    print(f"y_test shape: {y_test.shape}")
    
    # Build and train the LSTM model
    model, history, train_predictions, test_predictions, actual_y_train, actual_y_test, test_rmse = build_lstm_model(
        X_train, y_train, X_test, y_test, scaler, epochs=10
    )
    
    # Visualize the model performance
    prediction_df, forecast_df = visualize_predictions(df, test_dates, actual_y_test, test_predictions, ticker, test_rmse,model)    
    
    print("\nStock Price Prediction Completed!")
    print(f"Model RMSE: ${test_rmse:.2f}")
    
    return model, df, prediction_df, forecast_df

if __name__ == "__main__":
    main()
```
